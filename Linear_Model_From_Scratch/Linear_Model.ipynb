```python
from fastai.data.transforms import RandomSplitter
trn_split,val_split=RandomSplitter(seed=42)(df)
trn_indep,val_indep = t_indep[trn_split],t_indep[val_split]
trn_dep,val_dep = t_dep[trn_split],t_dep[val_split]
len(trn_indep),len(val_indep)
```

```python
def init_coef():
    torch.manual_seed(42)
    n_coef = trn_indep.shape[1]
    return torch.rand(n_coef).requires_grad_()
```

```python
def calc_pred(t_indep, coeffs):
    return (t_indep * coeffs).sum( axis = 1)

def calc_loss(t_indep, t_dep, coeffs):
    return torch.abs(calc_pred(t_indep, coeffs)-t_dep).mean()
```

```python
def step(coefficients, lr):
    loss = calc_loss(trn_indep,trn_dep, coefficients)
    loss.backward()
    print("dasjdijsad",coefficients.grad)
    with torch.no_grad():
        coefficients.sub_(coefficients.grad * lr)
        coefficients.grad.zero_()
    print(loss, " ; ")
```

```python
def train_model(epochs, lr, trn_indep, trn_dep):
    coefs = init_coef()
    for i in range(0,epochs):
        loss = calc_loss(trn_indep, trn_dep, coefs)
        loss.backward()
        with torch.no_grad():
            coefs.sub_(coefs.grad * lr)
            coefs.grad.zero_()
        print(loss)
    return coefs
```
